# Author Network

```python
# read from database

flag = False
authors = {}
papers = {}
names = {}
with open("../data/author_coauthor_paper_map.csv", "r") as f:
    for line in f:
        if flag:
            segs = line.split(",")
            aid = segs[0]
            coid = segs[1]
            pid = segs[2]
            name = segs[3].strip()[1:-1]
            
            # record name
            name = segs[3].strip()[1:-1]
            names[aid] = name
            
            # record co-authors
            if aid in authors.keys():
                authors[aid].add((coid, pid))
            else:
                authors[aid] = set([(coid, pid)])
            
            # record paper and its authors
            if pid in papers.keys():
                papers[pid].add(aid)
                papers[pid].add(coid)
            else:
                papers[pid] = set([aid, coid])
        else:
            flag = True
```

## 2. Author with largest number of coauthors

```python
unique_coauthors = {}

for aid in authors.keys():
    coauthors = set()
    for pair in authors[aid]:
        coauthors.add(pair[0])
    unique_coauthors[aid] = coauthors
```

```python
num_co = {}
for au in unique_coauthors.keys():
    num_co[au] = len(unique_coauthors[au])
```

```python
import operator

sorted_x = sorted(num_co.items(), key=operator.itemgetter(1))
```

```python
sorted_x.reverse()
```

```python
names = [names[m[0]] for m in sorted_x[0:10]]
```
