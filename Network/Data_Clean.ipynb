# Data Clean

## 1. Identify duplicate names

```python
# read data

flag = False

# mapping from author_id to name
mapping = {}

with open("../data/author_coauthor_paper_map.csv", "r") as f:
    for line in f:
        if flag:
            segs = line.split(",")
            aid = segs[0]
            name = segs[3].strip()[1:-1]
            mapping[aid] = name
        else:
            flag = True
```

```python
# check if any pair too similar
import Levenshtein

keys = mapping.keys()
for i in range(0, len(keys)):
    for j in range(i+1, len(keys)):
        first = keys[i]
        second = keys[j]
        dist = Levenshtein.distance(mapping[first], mapping[second])
        if dist <=2:
            print first+","+second+",",mapping[first],",",mapping[second]
```

## 2. Generate name dependencies

```python
dependency = {}
with open("duplicates.txt", "r") as f:
    for line in f:
        segs = line.split(","):
        id_a = segs[0]
        id_b = segs[1]
        if id_a in dependency.keys():
            dependency[id_a].append(id_b)
        else:
            dependency[id_a] = [id_b]
```
