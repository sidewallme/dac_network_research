```python
import rake
```

```python
import operator
```

```python
rake_object = rake.Rake("SmartStoplist.txt", 3, 3, 1)
```

```python
titles = []
abstracts = []

with open("/Users/erichsu/Documents/research/dac_network_research/Data/paper_abstract_years/paper_abstract_years_2014.csv", "r") as f:
    for line in f:
        seg = 0
        seg = line.find('","')
        title = line[1:seg]
        titles.append(title)
        
        abstract = line[seg+3:-7]
        abstracts.append(abstract)
```

```python
title_string = ". ".join(titles)
abstract_string = " ".join(abstracts)
```

```python
title_keywords = rake_object.run(title_string)
```
